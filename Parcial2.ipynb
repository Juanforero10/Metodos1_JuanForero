```python
M = np.array([[2,1],[0.5,0.3]])
print(M)
```

```python
 def temperatura(parametros,x,y):
        
        M,N = parametros.shape
        
        suma = 0
        sumaj = 0
        
        for i in range(M):
            
            sumai = 0
            
            for j in range(N):
                
                
                
                sumaint = parametros[i][j]*(x**i)*(y**j)
                sumaj += sumaint
                sumai += sumaj
                sumaj = 0
            
            suma += sumai 
            
            
            
        return(suma)
        print(suma)
    
            
        
            
            
```

```python
 temperatura(M,0,0.5)
```
