```python
x_0 = 5
a = 25214903917
c = 11
m = 2**48



```

```python
x = x_0
n = 1
k = 1

valores_y = []
valores_x = range(1,31,1)


while k <= 30:
    suma = 0
    while n <= cant_ale:
        x_i = x/m
        contador = 0
        while contador < k:
            x = (((a*x)+c)%m)
            contador +=1
        x_k = x/m
        x_ik = x_i*x_k
        suma += x_ik
        x = x_i*m
        x = (((a*x)+c)%m)
        
        n += 1
    print("C_" + str(k) + " = " + str(suma/(cant_ale)))
    valores_y.append(suma/(cant_ale))
    k += 1
    x = x_0
    n = 1

plt.ylim(0,0.5)
plt.xlabel("k-ésimo")
plt.ylabel("C(k)")
plt.plot(valores_x,valores_y)
    
        
            
```
