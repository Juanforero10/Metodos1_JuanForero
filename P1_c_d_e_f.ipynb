### Derivada progresiva

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
h = 0.01
N = int(20/h)+1
print(N)
```

```python
x = np.linspace(0.1,1.1,N)
function = lambda x: np.sqrt(np.tan(x))
```

```python
def GetDerivative(x,f,h):
    return (f(x+h)-f(x))/(h)
```

```python
Derivative = GetDerivative(x,function,h)
```

```python
plt.plot(x,Derivative)
```

```python
ExactDf = lambda x: ((1/np.cos(x))**2)/(2*np.sqrt(np.tan(x)))
ExactDerivative = ExactDf(x)
```

```python
plt.plot(x,Derivative,color='r',label='Right Derivative')
plt.plot(x,ExactDerivative,color='b', label='Exact Derivative')
plt.legend()
```

```python
plt.plot(x,np.abs(Derivative-ExactDerivative))
```

### Derivada central

```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
h = 0.01
N = int(20/h)+1
print(N)
```

```python
x = np.linspace(0.1,1.1,N)
function = lambda x: np.sqrt(np.tan(x))
```

```python
def GetDerivative(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
Derivative = GetDerivative(x,function,h)
```

```python
plt.plot(x,Derivative)
```

```python
ExactDf = lambda x: ((1/np.cos(x))**2)/(2*np.sqrt(np.tan(x)))
ExactDerivative = ExactDf(x)
```

```python
plt.plot(x,Derivative,color='r',label='Right Derivative')
plt.plot(x,ExactDerivative,color='b', label='Exact Derivative')
plt.legend()
```

```python
plt.plot(x,np.abs(Derivative-ExactDerivative))
```

### La derivada central tiene una mayor precisión, ya que para todos los puntos en x, el error nodal de la derivada central es menor que el error nodal de la derivada progresiva. Por ejemplo, para x = 1, el error es aprox. 0.015 para la progresiva y 0.00025 para la central. Lo anterior se cumple para todo x.
